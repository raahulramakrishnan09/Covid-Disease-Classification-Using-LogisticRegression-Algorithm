```python
#Using_dropna_to_remove_the_rows_with_missing_Values
df.dropna(inplace=True)
```

```python
#Replacing_several_values
df['CLASIFFICATION_FINAL'].replace([2,3],0,inplace=True)
df['CLASIFFICATION_FINAL'].replace([4,5,6,7],1,inplace=True)
```

```python
#Dropping_the_column
df=df.drop(['DATE_DIED','PATIENT_TYPE','AGE','HIPERTENSION','TOBACCO'],axis=1)
```

```python
#Heatmap
plt.figure(figsize=(20,15))
corr=df.corr()
sns.heatmap(corr,annot=True,fmt='.2g',linewidth=1)
plt.show()
```

```python
x=df.drop('CLASIFFICATION_FINAL',axis=1)
y=df['CLASIFFICATION_FINAL']
```

```python
#Logistic_Regression_Algorithm
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
```

```python
reg.fit(x,y)
```

```python
#Score
reg.score(x,y)
```

```python
#y=mx+c
#coefficient(m)
reg.coef_
```

```python
#intercept(c)
reg.intercept_
```

```python
#training_and_testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
x_train
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,r2_score,mean_absolute_error,mean_squared_error
#testing_accuracy_score
accuracy_score(y_pred,y_test)
```

```python
#confusion_matrix
confusion_matrix(y_pred,y_test)
```

```python
#r2_score
r2_score(y_pred,y_test)
```

```python
#mean_absolute_error
mean_absolute_error(y_pred,y_test)
```

```python
#mean_squared_error
mean_squared_error(y_pred,y_test)
```

```python
#root_mean_squared_error
np.sqrt(mean_squared_error(y_pred,y_test))
```
